### Question 1
Create a dataset of the population of a city in thousands for the past 10 years and create a histogram graph to visualize the population trend over time.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

popu=[13072,13228,13386,13690,13831,13541,13963,14075,14171,14286]
# year=np.arange(2013,2023)
year=['2013','2014','2015','2016','2017','2018','2019','2020','2021','2022']
#print(year)
plt.bar(year,popu)
plt.xlabel('Year')
plt.ylabel('Population in Thousands')
plt.title('Population trend over last 10 years of Jalgoan(in thousands)')
plt.show()
```

### Question 2

Plot the function $y = x^2 - 10$ for $x \in [-10, 10]$ with intervals of $0.1$.


```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(-10,10.1,0.1)
#print(x)
plt.plot(x,(x**2)-10,label='x vs x**2-10 graph')
plt.xlabel('value of x [-10,10]')
plt.ylabel('x**2-10')
plt.title('Graph of x**2-10 where x=[-10,10]')
plt.show()
```

### Loading Dataset for Questions 3 and 5

```python
'''
Run this cell before attempting questions 3 and 5
Also it would be better to explore the data before jumping to the question
'''
# dataset loading for Question 3 and Question 5
from sklearn.datasets import load_digits

digits_ds = load_digits()
# digit images in an array
images = digits_ds.images
'''
`image_labels` corresponding to images array
above `images` array such that 
for any index i `image_labels[i]` contains label for `image[i]` 
'''
image_labels = digits_ds.target
```

### Question 3
Using the `image_labels`, make a pie chart and horizontal bar graph giving the frequency of data points (labels) for each digit. Plot both pie chart and bar graph in the same plot.



```python
import numpy as np
a=np.unique(image_labels)

listx=[]
for i in range(len(a)):
  listx.append(i)

import matplotlib.pyplot as plt

plt.subplot(211)
plt.pie(a,labels=listx)
plt.title('Pie Plot')

plt.subplot(212)
plt.bar(listx,a)
plt.title('Bar Plot')
plt.show()
```

### Question 4
Suppose you have a dataset of stock prices for 4 different companies over the past year. You want to create a figure with 4 subplots, each showing the closing price of one of the companies over time.

Also plot the same in the same graph, with different colors each representing a company .

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

month=np.arange(1,13)
#print(month)
# print('GRAPH 1')

#TATA
tata=[452.10,387.95,439.40,412.75,404.60,471.10,449.60,411.80,443.55,437.60,433.75,454.05]
plt.subplot(511)
plt.plot(month,tata,'blue')
plt.xlabel('Months of year 2022')
plt.ylabel('Share Closeing Price')
plt.title('TATA Share Price')
plt.show()

#ADANI
adani=[299.55,332.75,334.90,373.50,413.95,313.95,262.85,323.25,280.35,185.10,123.75,106.25]
plt.subplot(512)
plt.plot(month,adani,'orange')
plt.xlabel('Months of year 2022')
plt.ylabel('Share Closeing Price')
plt.title('Adani Share Price')
plt.show()

#WIPRO
wipro=[392.75,406.90,386.55,394.25,413.55,423.70,416.05,478.05,508.80,591.90,555.80,572.60]
plt.subplot(513)
plt.plot(month,wipro,'red')
plt.xlabel('Months of year 2022')
plt.ylabel('Share Closeing Price')
plt.title('Wipro Share Price')
plt.show()

#NDTV
ndtv=[248.10,345.80,446.30,321.75,373.05,471.50,256.60,176.00,170.50,189.95,227.00,149.25]
plt.subplot(514)
plt.plot(month,ndtv,'green')
plt.xlabel('Months of year 2022')
plt.ylabel('Share Closeing Price')
plt.title('NDTV Share Price')
plt.show()


#COMMON GRAPH
# print('GRAPH 2')
fig=plt.figure(figsize=(6,20))
plt.subplot(515)
plt.plot(month,tata,label='TATA',color='blue')
plt.plot(month,adani,label='Adani',color='orange')
plt.plot(month,wipro,label='Wipro',color='red')
plt.plot(month,ndtv,label='NDTV',color='green')
plt.xlabel('Months of year 2022')
plt.ylabel('Share Closeing Price')
plt.title('Plot of 4 different Companies share price')
plt.legend()
plt.show()
```
